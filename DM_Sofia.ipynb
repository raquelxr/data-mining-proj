#### EnrollmentStart / EnrollmentFinish

We start by turning all dates into datetime format:

```python
data['EnrollmentStart'] = pd.to_datetime(data['EnrollmentStart'])
data['EnrollmentFinish'] = pd.to_datetime(data['EnrollmentFinish'])

data['LastPeriodStart'] = pd.to_datetime(data['LastPeriodStart'])
data['LastPeriodFinish'] = pd.to_datetime(data['LastPeriodFinish'])

data['DateLastVisit'] = pd.to_datetime(data['DateLastVisit'])
```

```python
data[data['EnrollmentStart'] == data['EnrollmentFinish']]['LastPeriodFinish'].value_counts()
```

Most People have LastPeriodFinish = '2019-12-31' (as expected for renovating members), so we'll look into the other cases:

##### Members who have not been active in Current Period:

We'll check the case of people who haven't visited the Facility since the last Period:

```python
data.loc[data['DateLastVisit']< '2019-06-30', 'Dropout'].value_counts()
```

The people who had DateLastVisit < '2019-06-30' and don't have a DropOut are the poeple who have 'EnrollmentStart'= 'EnrolmentFinish', which tells us that:
- these people have dropped out but that has not been added to the system.
- Their contract is still active as they've been paying, but not comming to the gym.

```python
data[data['EnrollmentStart'] == data['EnrollmentFinish']].loc[data['DateLastVisit']< '2019-06-30']
```

----

Some people have DateLastVisit in their LastPeriod (233):
- Those people will be considered the ones who Droped Out - EnrollmentFinished on the date of LastVisit.
- The rest will be considered active.

```python
data[data['EnrollmentStart'] == data['EnrollmentFinish']][data['DateLastVisit']< data['LastPeriodFinish']][data['DateLastVisit']> data['LastPeriodStart']][data['DateLastVisit']< '2019-06-30']
```

```python
index_i =[]
index_drop = []
for i in data.index:
    start = data['LastPeriodStart'][i]
    end = data['LastPeriodFinish'][i]
    date_visit= data['DateLastVisit'][i]
    if data['EnrollmentFinish'][i] == data['EnrollmentStart'][i]:
        if start<=date_visit <=end:
            if data['DateLastVisit'][i] < pd.Timestamp(dt.date(2019,6,30)):
                index_i.append(i)
```

These people are the Drop Outs:

```python
data.loc[index_i, 'Dropout']=1
```

```python
data.loc[index_i, 'EnrollmentFinish'] = data.loc[index_i, 'DateLastVisit']
```

-----

We now need to check our non-drop out members:

```python
data[data['EnrollmentStart'] == data['EnrollmentFinish']][data['DateLastVisit']< '2019-06-30']['LastPeriodFinish'].value_counts()
```

Members that don't have LastPeriodFinish == '2019-12-31' will be removed (it should be '2019-12-31' so they'll be considered errors):

```python
data.drop(data[data['EnrollmentStart'] == data['EnrollmentFinish']][data['DateLastVisit']< '2019-06-30'][data['LastPeriodFinish']!= '2019-12-31'].index, inplace = True)
```

#### DateLastVisit

We need to check how many people don't have Period that matches:
- DateLastVisit
- EnrollmentFinish-2

```python
from dateutil.relativedelta import relativedelta

sum = 0
index_i =[]
for i in data.index:
    start = data['LastPeriodStart'][i]
    end = data['LastPeriodFinish'][i]
    date_test = data['EnrollmentFinish'][i] - relativedelta(months=2)
    date_visit= data['DateLastVisit'][i]
    if start<=date_test <=end:
        sum = sum
    elif start<=date_visit <=end:
        sum = sum
    else:
        sum+=1
        index_i.append(i)
print(sum)
```

```python
data_wrong =data.loc[index_i]
```

O restante dos clientes vão ser considerados erros e droped:

```python
sum = 0
index_drop =[]
for i in data.index:
    start = data['LastPeriodStart'][i]
    end = data['LastPeriodFinish'][i]
    date_test = data['EnrollmentFinish'][i] - relativedelta(months=2)
    date_visit= data['DateLastVisit'][i]
    if start<=date_test <=end:
        sum = sum
    elif start<=date_visit <=end:
        sum = sum
    else:
        sum+=1
        index_drop.append(i)
print(sum)
```
